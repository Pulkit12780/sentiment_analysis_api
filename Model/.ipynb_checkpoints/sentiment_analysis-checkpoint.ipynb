Data Preprocessing

```python
def clean_text(text):
    output = " ".join(text.split())
    return output
```

```python
df['review'] = df['review'].apply(clean_text)
```

```python
x = df['review']
y = df['sentiment']
```

```python
def convert_to_label(text):
    if text == 'positive':
        return 1
    else:
        return 0
```

```python
y = y.apply(convert_to_label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from keras.preprocessing.text import one_hot, Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, GlobalMaxPooling1D, Embedding, Conv1D, LSTM
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
```

```python
X_train_tokenised = tokenizer.texts_to_sequences(X_train)
X_test_tokenised = tokenizer.texts_to_sequences(X_test)
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
maxlen = 100

X_train_pad = pad_sequences(X_train_tokenised, padding='post', maxlen=maxlen)
X_test_pad = pad_sequences(X_test_tokenised, padding='post', maxlen=maxlen)
```

```python
sentences = [text.split() for text in X_train]
```

```python
from gensim.models import Word2Vec

embedding_dim = 100

w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=embedding_dim,
    window=5,
    min_count=1,
    workers=4
)
```

```python
sample = np.zeros((3,2))
```

```python
sample[0]
```

```python
vocab_size = len(tokenizer.word_index) + 1

embedding_matrix = np.zeros((vocab_size, embedding_dim))

for word, index in tokenizer.word_index.items():
    if word in w2v_model.wv:
        embedding_matrix[index] = w2v_model.wv[word]
```

```python
embedding_matrix
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Dense
from tensorflow.keras.layers import Flatten, GlobalMaxPooling1D, Embedding, Conv1D, LSTM
```

```python
model= Sequential()
embedding = Embedding(input_dim = vocab_size,
    output_dim = 100,
    embeddings_initializer='uniform',
    weights=[embedding_matrix],
    input_length= 100 , 
    trainable=False
)

model.add(embedding)

model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])

print(model.summary())
```

```python
model.fit(X_train_pad, y_train, batch_size=128, epochs=6, verbose=1,validation_data = (X_test_pad, y_test) )
```

```python
score = model.evaluate(X_test_pad, y_test, verbose=1)
```

```python
print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```

```python
model1 = Sequential()
embedding = Embedding(input_dim = vocab_size,
    output_dim = 100,
    embeddings_initializer='uniform',
    weights=[embedding_matrix],
    input_length= 100 , 
    trainable=False
)

model1.add(embedding)
model1.add(LSTM(128))

model1.add(Dense(1, activation='sigmoid'))

model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
print(model1.summary())
```

```python
model1.fit(X_train_pad, y_train, batch_size=128, epochs=6, verbose=1,validation_data = (X_test_pad, y_test) )
```

```python
score = model1.evaluate(X_test_pad, y_test, verbose=1)
print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```

```python
def predict_sentiment(sample):
    clean_sample = [clean_text(sample)]
    clean_sample_tokenised = tokenizer.texts_to_sequences(clean_sample)
    clean_sample_padded = pad_sequences(clean_sample_tokenised, padding='post', maxlen=maxlen)
    prediction = model1.predict(cleaned_sample_padded)*10
    return prediction
```
